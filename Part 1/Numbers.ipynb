# Kalkulator systemów liczbowych dec -> hex, oct . . . ( do 36 )

```python
def calc(n: int, base: int) -> str:
    m = int()
    digits = list()
    map = '0123456789ABCDEFGHIJKLMNOPQRSTUWXYZ'

    while n > 0:
        if n / base < 1:
            digits.append(n)
            break
        elif n / base == 1: 
            digits.append(int(n / base))
            break
        # or divmode(n, b) -> (n, m)
        m = n % base
        n = n // base
        digits.append(m)

    return ''.join([map[x] for x in digits[::-1]])
```

```python
calc(173, 2)
```

```python
calc(65354, 16)
```

# Rounding ( Zaokrąglenia )

```python
from math import copysign
```

```python
def rounding(x: float) -> int: # Rounding AWAY from 0
    return copysign(1, x) * int(abs(x)+0.5)
```

```python
rounding(9.9)
```

```python
rounding(10.5)
```
